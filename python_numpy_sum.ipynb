```
import numpy as np
```

```
# 3 dimension array
arr = np.arange(0, 4*2*4)
arr
```

```
# 차원 변환
v = arr.reshape([4, 2, 4]) # row, col, depth
v
# 0, 8, 16, 24: row
# 0, 4: col
# 0, 1, 2, 3: depth
```

```
v.ndim
# sum 함수의 axis 파라미터의 기본값은 “None”입니다.
# axis는 대상 데이터의 Dimension보다 작은 값을 설정합니다.
# 예제 데이터의 Dimension은 3이기 때문에 axis는 2까지 설정 가능합니다.
```

# axis=None

```
v.sum() # scalar
```

# axis =0

```
res01 = v.sum(axis=0) # row sum
print(res01.shape) # 4*2*4 -> 2*4
print(res01)
```

# axis=1

```
res02 = v.sum(axis=1)
print(res02.shape) # 4*2*4 -> 4 * 4
print(res02)
```

# axis=2

```
res03 = v.sum(axis=2)
print(res03.shape) # 4*2*4 -> 4*2
print(res03)
```
